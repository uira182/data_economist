# Teste: ibge.get(t, n, v, p, c)

Testa a função **ibge.get()** que monta a URL da API SIDRA a partir dos parâmetros:
- **t** = tabela
- **n** = região (tuple: nível, valor), ex. (3, "all") → n3/all
- **v** = variável
- **p** = período ("all", "first", "last")
- **c** = classificação (id_class, id_cat), ex. (544, 129317)

Cada célula tenta uma combinação e exibe o resultado ou o erro tratado.

```python
import os
import sys
if os.path.basename(os.getcwd()) == "tests":
    os.chdir("..")
raiz = os.path.abspath(".")
src = os.path.join(raiz, "src")
if os.path.isdir(src) and src not in sys.path:
    sys.path.insert(0, src)
for key in list(sys.modules.keys()):
    if key == "data_economist" or key.startswith("data_economist."):
        del sys.modules[key]
from data_economist import ibge
import requests
print("OK — ibge importado.")
```

## 1) Só tabela: ibge.get(t=8888)

Mínimo: apenas o número da tabela. A API pode exigir mais parâmetros.

```python
try:
    dados = ibge.get(t=8888)
    print(f"Sucesso. Tipo: {type(dados).__name__}, tamanho: {len(dados) if isinstance(dados, list) else 'N/A'}")
    if isinstance(dados, list) and len(dados) > 0:
        print(f"Primeiro elemento (amostra): {str(dados[0])[:200]}...")
    dados
except requests.HTTPError as e:
    print(f"ERRO HTTP: {e.response.status_code}")
    print(f"URL: {e.response.url}")
    print(f"Corpo: {e.response.text[:500] if e.response.text else '(vazio)'}")
except Exception as ex:
    print(f"ERRO: {type(ex).__name__}: {ex}")
```

## 5) Combinação completa: tabela + região + variável + período + classificação

## 6) Teste de erro: tabela inexistente

Esperado: API devolve erro (ex.: 404 ou 400); exibimos o tratamento.

```python
try:
    dados = ibge.get(t=99999999)
    print("Sucesso (inesperado):", type(dados))
    dados
except requests.HTTPError as e:
    print(f"ERRO HTTP (esperado): {e.response.status_code}")
    print(f"URL: {e.response.url}")
    print(f"Corpo: {e.response.text[:500] if e.response.text else '(vazio)'}")
except Exception as ex:
    print(f"ERRO: {type(ex).__name__}: {ex}")
```

## 7) Resumo: URL gerada por ibge.get

Para conferir a URL que seria chamada (sem fazer o GET), use os metadados ou monte manualmente. Aqui chamamos get e mostramos o resultado em DataFrame.

```python
import pandas as pd
try:
    dados = ibge.get(t=8888, n=(3, "all"), v=12606, p="last", c=(544, 129317))
    if isinstance(dados, list) and len(dados) > 1:
        df = pd.DataFrame(dados[1:])
        print(f"DataFrame: {len(df)} linhas × {len(df.columns)} colunas")
        df
    else:
        print("Resposta:", dados)
except requests.HTTPError as e:
    print(f"ERRO HTTP: {e.response.status_code}")
    print(f"URL: {e.response.url}")
    print(f"Corpo: {e.response.text[:300]}")
except Exception as ex:
    print(f"ERRO: {type(ex).__name__}: {ex}")
```
