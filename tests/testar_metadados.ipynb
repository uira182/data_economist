## Obter metadados da tabela 8888

A tabela 8888 é a **Produção Física Industrial** (PIMPF). O resultado é um dicionário com id, nome, URL, pesquisa, assunto, periodicidade, nivelTerritorial, variaveis e classificacoes.

```python
from data_economist import ibge

meta = ibge.metadados(8888)
print("Chaves:", list(meta.keys()))
print()
print("Nome:", meta["nome"])
print("URL:", meta["URL"])
print("Pesquisa:", meta["pesquisa"])
print("Periodicidade:", meta["periodicidade"])
```
